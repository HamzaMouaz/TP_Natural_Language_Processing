```python
import spacy
import gensim
from gensim import corpora
from gensim import models
import os
import nltk
nltk.download("gutenberg")
nltk.download('reuters')
from nltk.corpus import gutenberg
from nltk.corpus import reuters
from nltk.tokenize import word_tokenize
from gensim.models import Phrases



# Obtenez la liste des identifiants de documents dans le corpus Reuters
document_ids = reuters.fileids()

# Créez une liste vide pour stocker les documents
documents = []
mots_vides= nlp.Defaults.stop_words
# Parcourez les identifiants de documents et lisez le texte de chaque document
for doc_id in document_ids[:10]:  # Vous pouvez spécifier le nombre de documents que vous souhaitez traiter
    document = reuters.raw(doc_id)
    text = []
    doc = nlp(document)
    #print(doc)
    for w in doc:
        if not w.is_stop and not w.is_punct and not w.like_num and (w.text != "\n"):
            text.append(w.text)
    documents.append(text)
    #documents.append(word_tokenize(document))

print("Mots Normales ##########################")
print(documents)

#bigrams
bigram = gensim.models.Phrases(documents, threshold=1, min_count=3)
textss = [bigram[line] for line in documents]

print("Mots avec 2-Grams ##########################")

print(f"{textss}")

#print(textss)

#trigrams
trigram = gensim.models.Phrases(textss, threshold=1, min_count=3)
textsss = [trigram[line] for line in textss]

print("Mots avec 3-Grams ##########################")
print(textsss)
print("aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa")

dictionary = corpora.Dictionary(textss)
print(dictionary.token2id)

```
